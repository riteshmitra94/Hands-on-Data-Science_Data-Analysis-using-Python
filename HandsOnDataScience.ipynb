How many times each product appears in DataSet, we get the count

```python
s = df['Product Name']

s.value_counts(dropna=False)
```

Aggregation by Year

```python
df['Month of Order Date'].str[:4]
```

Describe the aggregate by Year

```python
df.groupby(df['Month of Order Date'].str[:4]).describe()
```

Renaming the column to more appropriate one

```python
df.groupby(df['Month of Order Date'].str[:4]).describe().reset_index().rename(columns={"Month of Order Date":"Year"})
```

Adding Product too and renaming column to Year

```python
yearly_product_sales = df.groupby([df['Month of Order Date'].str[:4], "Product Name"]).sum().reset_index()
```

```python
yearly_product_sales.rename(columns={"Month of Order Date":"Year"})
```

Finding overall Product's overall Sales

```python
df.groupby("Product Name").sum().reset_index()
```

Exporting the data back to CSV and Excel

```python
yearly_product_sales.to_csv("yearly_product_sales.csv", header='True', index=False, encoding='utf-8')
```

```python
yearly_product_sales.to_excel("yearly_product_sales.xlsx", header='True', index=False)
```
